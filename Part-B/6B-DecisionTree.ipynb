```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(['Class'],axis=1),df['Class'],test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')
model = dtc.fit(x_train,y_train)
prediction = model.predict(x_test)
print("prediction : ",prediction[:5])
```

```python
from sklearn.metrics import confusion_matrix, matthews_corrcoef, classification_report, accuracy_score
import seaborn as sns
cm = confusion_matrix(y_test,prediction)
print("Accuracy = ",accuracy_score(y_test,prediction))
print("matthews_corrcoef = ",matthews_corrcoef(y_test,prediction))
print(classification_report(y_test,prediction))
sns.heatmap(cm,annot = True)
```

```python
from sklearn import tree
tree.plot_tree(dtc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_gini = DecisionTreeClassifier(criterion='gini')
model = dtc_gini.fit(x_train,y_train)
prediction_gini = model.predict(x_test)
print("prediction : ",prediction_gini[:5])
```

```python
from sklearn.metrics import confusion_matrix, matthews_corrcoef, classification_report, accuracy_score
import seaborn as sns
cm = confusion_matrix(y_test,prediction_gini)
print("Accuracy = ",accuracy_score(y_test,prediction_gini))
print("matthews_corrcoef = ",matthews_corrcoef(y_test,prediction_gini))
print(classification_report(y_test,prediction_gini))
sns.heatmap(cm,annot = True)
```

```python
tree.plot_tree(dtc_gini)
```
